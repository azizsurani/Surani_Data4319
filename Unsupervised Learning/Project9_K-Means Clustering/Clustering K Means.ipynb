```python
from pandas import read_csv #  imports a CSV file to DataFrame format.
dataset = read_csv('online_shoppers_intention.csv')# read csv
```

```python
dataset['Month'] = dataset['Month'].astype('category')
dataset['VisitorType'] = dataset['VisitorType'].astype('category')
dataset['Weekend'] = dataset['Weekend'].astype('category')
dataset['Revenue'] = dataset['Revenue'].astype('category')

dataset.dtypes
```

```python
dataset['Month'] = dataset['Month'].cat.codes
dataset['VisitorType'] = dataset['VisitorType'].cat.codes
dataset['Weekend'] = dataset['Weekend'].cat.codes
dataset['Revenue'] = dataset['Revenue'].cat.codes
```

```python
data = dataset.drop_duplicates()
```

```python
data.to_csv("online_shoppers_intention.csv", index=False)
```

```python
from pandas import read_csv #  imports a CSV file to DataFrame format.
dataset = read_csv('online_shoppers_intention.csv')# read csv
```

```python
f,ax = plt.subplots(figsize=(9, 9))
sns.heatmap(dataset.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
dataset.head()
```

```python
x = dataset.copy()
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python

kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
cluster_new = x.copy()
cluster_new['cluster_pred'] = kmeans_new.fit_predict(x_scaled)
cluster_new
```

```python
plt.scatter(cluster_new['ProductRelated'],cluster_new['Revenue'], c = cluster_new['cluster_pred'], cmap = 'rainbow')
plt.xlabel('ProductRelated')
plt.ylabel('Revenue')
plt.show()
```

```python
c = cluster_new['cluster_pred']
dataset['c'] = c
```

```python
dataset.to_csv("clus.csv", index=False)
```
